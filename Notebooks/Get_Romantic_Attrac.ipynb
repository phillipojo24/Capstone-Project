```python
# default tripadvisor website of attractions
url = "https://www.tripadvisor.com"
```

```python
from webdriver_manager.chrome import ChromeDriverManager

city_recomendations = ["Big Sur, California", "Ellensburg, Washington", "Healdsburg, California", "Hood River, Oregon", "Walla Walla, Washington", "San Luis Obispo, California", "Seabrook, Washington", "Astoria, Oregon", "Coronado, California", "The Dalles, Oregon","McMinnville, Oregon", "LA Jolla, California","Cannon Beach, Oregon","Palm Springs, California","Ventura, California","Santa Barbara, California","Fort Bragg, California","Nevada City, California","Sausalito, California","Truckee, California","Two Harbors, Catalina Island, California","Langley on Whidbey Island, Washington", "Hartford, Connecticut", "New Haven, Connecticut", "Killingworth, Connecticut", "Dorchester County, Maryland", "Harford County, Maryland", "Nantucket, Massachusetts", "Point Pleasant, New Jersey", "Cape May, New Jersey", "Sullivan County , New York", "Seneca Falls, New York", "Buffalo, New York", "Syracuse, New York", "Beacon, NY", "Saratoga Springs, New York", "Carlisle, PA", "Hershey, PA", "Philadelphia,PA", "Manchester, Vermont", "Middlebury, Vermont", "Orange County, Virginia", "Harrisonburg, Virginia", "Richmond, Virginia"]


#Name the csv file
filename = "Romantic.csv"

#Coloumn name for csv file
headers = "Attraction,Address,Rank,Rating,Number_of_Reviews,Label\n"

#Open file and write
f = open(filename, 'w')
f.write(headers)
```
